```python
from utils import DATA_DIR
import pandas as pd
```

```python
ds_small_labeled1_path = DATA_DIR / "datasets" / "small" / "agreed.csv"
ds_small_labeled2_path = DATA_DIR / "datasets" / "small" / "agreed2.csv"
ds_small1 = pd.read_csv(ds_small_labeled1_path, sep="\t", dtype={"label": int})
ds_small2 = pd.read_csv(ds_small_labeled2_path, sep="\t", dtype={"label": int})
```

```python
ds_small1.drop(columns=["segment_id", "annotator_1", "annotator_2"], inplace=True)
ds_small2.drop(
    columns=["segment_id", "annotator_1", "annotator_2", "annotator_3"], inplace=True
)
# Merge the two (they have the same columns after dropping)
ds_small = pd.concat([ds_small1, ds_small2], ignore_index=True)
ds_small.rename(columns={"agreed_rating": "label"}, inplace=True)

# Drop rows without label
ds_small = ds_small.dropna(subset=["label"])

ds_small.to_parquet(DATA_DIR / "datasets" / "small" / "agreed.parquet", index=False)
```
